```python
import pandas as pd
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

- **X**: `Age`와 `EstimatedSalary`
- **y**: `Purchased`
- `Age`(나이)와 `EstimatedSalary`(연봉추정치) 값을 통해 `Purchased`(구입여부)를 판단하는 **Logistic Regression** model을 생성

```python
X = dataset.drop(columns=['Purchased'])
y = dataset['Purchased']
```

- **Training data**와 **Test data**를 **7:3**으로 분리

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

- `X_train` **Scaling**
> - **Scaling**이란 `X`에 대한 각 특징의 크기를 유사한 범위로 조정하는 단계임
> - 예를 위 `X`의 경우, `Age`는 **18~60**, `EstimatedSalary`는 **15000~150000**의 값을 가지며, **Machine learning** 종류에 따라 각 특징 크기 범위가 크게 차이날 때, 성능이 매우 나빠질 수 있음 (ex> KNN, SVM, Ridge regression 등등)
- 대표적인 **Scaling** 방법
> - **StandardScaler**: `X`의 모든 특징을 **평균 0**, **표준편차(분산) 1**인 값으로 변환함, **0**에 가까울 수록 평균에 근사한 값, 평균보다 작은 값은 음수, 큰 값은 양수가 됨
> - **MinMaxScaler**: `X`의 모든 특징을 지정된 범위의 값으로 변환함 (ex> `0~1`)
- **sklearn**의 **Scaler** 사용 방법
> - `Scaler.fit(X)`: 입력된 `X`의 정보를 통해 변환식을 찾음, **StandardScaler**의 경우 **평균과 표준편차**, **MinMaxScaler**의 경우 **최소/최대값**을 찾아 변환식을 찾음
> - `Scaler.transform(X)`: `fit`을 수행하여 찾은 변환식을 통해 실제 `X`를 변환함


```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# sc.fit(X_train): X_train을 기반으로 X 변환식을 찾음
sc.fit(X_train)
# sc.transform(X_train): X_train을 실제로 변환함
Xsc_train = sc.transform(X_train)
# StandardScaler 변환을 수행하면, 일반적으로 값이 -3~3사이의 값을 갖는다.
```

- **Logistic Regression** 학습

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
# Scaling을 적용한 Xsc_train으로 학습한 모델
lr_sc = LogisticRegression()
lr_sc.fit(Xsc_train,y_train)
```

- **X_test**에 대한 **Logistic regression** model의 예측값

```python
y_pred = lr.predict(X_test)
```

```python
# Scaling을 적용한 Xsc_test로 학습한 모델
# 이 때, X_test는 X_train을 Xsc_train로 변환할 때 사용한 변환식을 그대로 사용해야 함
# 즉, X_test에 대한 fit은 수행하지 않음
# Scaling, Encoding, Vectorizer(데이터의 변환) 및 Machine learning model(classifier, regressor 등)의 fit은 항상 학습데이터를 적용하며
# fit에는 테스트 데이터를 사용하지 않는 것을 원칙으로 함
Xsc_test = sc.transform(X_test)
ysc_pred = lr_sc.predict(Xsc_test)
```

- **Logistic regression** model의 성능 평가
> - **Accuracy**, **Precision**, **Recall**, **Confusion Matrix**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
# [[TN, FP]
#  [FN, TP]]
print(precision_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
```

```python
print(accuracy_score(y_test,ysc_pred))
print(confusion_matrix(y_test,ysc_pred))
print(precision_score(y_test,ysc_pred))
print(recall_score(y_test,ysc_pred))
```

- **Sklearn**의 **Logistic regression**의 경우, 성능이 **Scaling** 적용과 무관하므로 **Scaling**을 생략해도 됨
- **(주의)** 과거 **sklearn** 버전의 **Logistic regression**의 경우, **Scaling**이 필수일 수 있음

- `LogisticRegression.predict(X)`: class label **0 또는 1** 을 반환(확률이 0.5보다 클 경우 1, 작을 경우 0)
- `LogisticRegression.predict_proba(X)`: 각 class(0 또는 1)에 속할 **확률을 반환**, 각 데이터마다 2개의 값을 반환하며 각각 **class 0(y=0) 속할 확률과 class 1(y=1)에 속할 확률**을 의미함

```python
print(lr.predict(X_test))
print(lr.predict_proba(X_test))
```

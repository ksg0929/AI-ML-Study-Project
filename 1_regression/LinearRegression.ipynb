```python
import pandas as pd
```

- `pandas` 라이브러리는 테이블 형태의 데이터를 다루는 라이브러리이다.

- `pd.read_csv('file name')`: csv 형태의 파일을 로드한다.
- `dataset = pd.read_csv('Salary_Data.csv')`: Salary_Data.csv를 로드한 결과를 dataset이라는 이름으로 저장한다. 이 때, `read_csv`에 의해 로드된 데이터는 `DataFrame` 타입으로 저장된다.
- `DataFrame` 타입은, 행(`index`), 열(`columns`), 값(`values`)로 구성된 2차원 테이블 형태의 데이터이다.

## 1. Data Load

```python
dataset = pd.read_csv('Salary_Data.csv')
```

```python
dataset
```

```python
print(dataset.index)   # 행
print(dataset.columns) # 열
print(dataset.values)  # 행과 열의 이름을 제외한 순수한 값 (ndarray 타입)
```

- **Salary_Data**
> - `YearsExperience`(경력), `Salary`(연봉) 두 개의 열(column)으로 구성된 데이터
> - 총 30개의 행(데이터 샘플의 수)이 있음

- **Linear Regression**
> - `YearsExperience` 값을 입력하면, `Salary` 값을 예측하는 **Linear regression** model 생성
> - `X`: `YearsExperience`, `y`:`Salary`
> - 주어진 30개의 데이터 샘플 중, 일부는 학습데이터로 사용하고, 나머지는 성능 평가를 위한 테스트 데이터로 사용함


## 2. X와 y 분리
- `DataFrame`의 특정 열 선택: `DataFrame['열의 이름']`
- `DataFrmae`의 특정 열 제외: `DataFrame.drop(columns=['열의 이름'])`
- 위에서 '열의 이름'은 모두 복수로 설정할 수 있음
- `DataFrame['열의 이름']`을 통해 하나의 열만 추출하면, `Series` 타입을 반환함
- `DataFrame.drop(columns=['열의 이름'])`은 남아 있는 열의 수와 관계없이 항상 `DataFrame`을 반환함
- 그러므로 `dataset['Salary']`은 `Series`, `dataset.drop(columns=['YearsExperience'])`은 `DataFrame` 타입이 됨
- 머신러닝이나 딥러닝을 다루는 `sklearn`, `tensorflow` 등의 입력 `X`는 항상 2차원의 형태여야 하므로, 여기서 `X`는 `DataFrame` 타입만 허용됨

```python
print(type(dataset['Salary']))
print(type(dataset.drop(columns=['YearsExperience'])))
print(dataset['Salary'])
print(dataset.drop(columns=['YearsExperience']))
```

```python
# X는 DataFrame
# y는 Series
y = dataset['Salary']
X = dataset.drop(columns=['Salary'])
```

## 3. 학습데이터와 테스트데이터 분리
- `sklearn.model_selection` 모듈의 `train_test_split` 사용
- `train_test_split(dataset1,dataset2,test_size,random_state)`
> - 위 함수는 입력된 데이터를 **임의로 shuffle**한 후, 전달된 test_size에 따라 **학습 및 테스트 데이터를 분리**하여 반환함
> - `datset1`, `dataset2`: 분리하고자하는 데이터 (`dataset1` 하나만 전달해도 무방)
> - `test_size`: 입력된 데이터 중, 분리하고자 하는 테스트 데이터의 비율 (0~1)
> - `radom_state`: 동일한 결과로 고정
> - **반환 결과 순서**: `dataset1_train, dataset1_test, dataset2_train, dataset_test`


```python
# train_test_split 기능 추가
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

- `X`의 training(학습) 데이터, `X`의 test(테스트) 데이터, `y`의 training(학습) 데이터, `y`의 test(테스트) 데이터 순서로 반환함
- `X`와 `y`는 각각 30개의 sample이므로 테스트데이터의 수는 30%인 9개, 학습데이터의 수는 70%인 21개가 됨

## 4. 선형회귀(Linear Regression) 학습
- `sklearn.linear_model`의 `LinearRegression`을 이용
- `LinearRegression.fit(X,y)`: 전달받은 X와 y데이터를 이용하여 선형회귀모델을 학습함

```python
from sklearn.linear_model import LinearRegression
```

```python
# LinearRegression model 생성
lr_salary = LinearRegression()
# 생성된 model 학습
lr_salary.fit(X_train,y_train)
```

## 5. 생성된 선형회귀 모델로 연봉(Salary) 예측하기
- `LinearRegression.predict(X)`: 예측할 데이터를 전달하면 해당 데이터에 대한 예측값을 반환 <br> `predict`는 반드시 `fit`을 실행한 이후, 사용할 수 있음
- 이 때, `X`는 반드시 2차원 형태(`DataFrame` 등)
- 하나의 데이터 샘플을 전달할 경우, 이중 괄호를 사용(2차원이라는 의미) ex> [[10]]

```python
# 경력 4.5년에 대한 연봉 예측
lr_salary.predict([[4.5]])
```

```python
# X_test에 대한 연봉 예측값 (y_pred)
y_pred = lr_salary.predict(X_test)
```

```python
# y_test: X_test에 대한 실제 값
# y_pred: X_test에 대한 lr_salary 선형회귀모델의 예측 값
print(y_test)
print(y_pred)
```
